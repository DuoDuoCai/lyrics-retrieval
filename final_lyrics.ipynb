## functions about first stage retrieval

```python
from pathlib import Path
import pandas as pd
import re
import torch.nn.functional as F
import torch
from torch import Tensor
from transformers import AutoTokenizer, AutoModel
import faiss


def average_pool(last_hidden_states,
                 attention_mask):
    last_hidden = last_hidden_states.masked_fill(~attention_mask[..., None].bool(), 0.0)
    return last_hidden.sum(dim=1) / attention_mask.sum(dim=1)[..., None]

def clean_invisible(text):
    if not isinstance(text, str): return ""
    return re.sub(r"[\u200b\u200c\u200d\u200e\u200f\u202f\u205f\u2060\u00a0\u180e\u2005]", "", text)


def _embed(texts, tokenizer, model, batch_size=64, device="cpu"):
    """
    <texts>: list of texts to embed
    <tokenizer> and <model>: the biencoder model used to embed.
    For each piece of text, run a forward-pass, extract hidden layer
    of all tokens and mean-pool over the sequence to get an embedding.
    Then L2-normalized them.
    """
    out = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i : i+batch_size]
        tok= tokenizer(batch,max_length=512,
                          padding=True, truncation=True,
                          return_tensors="pt").to(device)
        with torch.no_grad():
            h = model(**tok).last_hidden_state          # [B,T,H]

        # h: (145, 512, 1024), tok["attention_mask"][...None]: (145,1024,1)
        # 145: batch size, 512: seq length

        # each song's mean pool over sequence:
        # for each of a specific song (one text is one song),
        # compute a 1024-dim for each word, and average over the words
        #  emb: torch.Size([145, 1024])
        emb = (h * tok["attention_mask"][..., None]).sum(1) \
              / tok["attention_mask"].sum(1, keepdim=True)
        emb = F.normalize(emb, p=2, dim=1)
        out.append(emb.cpu())
    # torch.cat(out): if there are N 145*1024 matrices,
    # then then results has shape (N*145, 1024)
    return torch.cat(out).numpy().astype("float32")

def generate_score(info, tokenizer, model,
                   num_queries, top_k=6,
                   passage_batch=256, query_batch=64,
                   use_gpu=True):
    """
    <info>: info[0 : num_queries] are queries(prefixed 'query:') afterwards
    are songs (prefixed 'passage:').

    Embed all queries and songs with <model>, then do a FAISS approximate
    search for each query, and for each query generate an array of size
    top_k representing the index of those retrieved songs in the search index.

    Returns:
        D: (num_queries, top_k) cosine scores.
        I: (num_queries, top_k) passage indices of those top k songs
    """
    device = "cuda" if (use_gpu and torch.cuda.is_available()) else "cpu"
    model = model.to(device)
    queries   = info[:num_queries]
    passages  = info[num_queries:]
    # print("inside generate score, the queries have length, ", len(queries))
    # print("inside generate score, the passages have length, ", len(passages))
    p_vecs = _embed(passages, tokenizer, model,
                    batch_size=passage_batch, device=device)
    q_vecs = _embed(queries,  tokenizer, model,
                    batch_size=query_batch,  device=device)

    dim   = p_vecs.shape[1]
    index = faiss.IndexFlatIP(dim)
    index.add(p_vecs)
    D, I = index.search(q_vecs, top_k)
    return D * 100.0, I+num_queries
```

## functions related to data preprocessing and evaluation metrics

```python
import traceback
import sys
N=512    # total number of songs in the database


# ---------------------------------------- function definitions ----------------------------------------
def convert_titles_to_labels(nested_list_titles, search_index_w_titles):
    '''
    <nested_list_titles>:  list of music titles, may contain string or list of
     strings.
     <search_index_w_titles>: extract which song is at which index.

    return an array of exactly the same "shape" as <nested_list_titles> but with
     the integer index/label of those titles. The index are found in search_index_w_titles.
    '''
    result=[]
    for one_answer in nested_list_titles:
        try:
            if isinstance(one_answer, str):
                result.append(search_index_w_titles.index(one_answer.lower()))
            elif isinstance(one_answer, list):
                ind_res = []
                for individual_answer in one_answer:
                    ind_res.append(search_index_w_titles.index(individual_answer.lower()))
                result.append(ind_res)
            else:
                print("shouldn't be here!")
                break
        except Exception as e:
             traceback.print_exc()

    return result

def create_corpus(df):
  corpus = df["Lyric"].tolist()
  m=len(corpus)

  for j in range(m):
      corpus[j]="passage: "+corpus[j]

  titles = merged_df["Title"].tolist()

  return corpus,titles

def average_precision_at_K(returned_labels, answer_labels, K=K1):
    '''
    <K>: when computing precision, you want to include top how many documents.
    Cannot exceed the global variable K1.
    <returned_labels>: number queries * K. An np array of retrieved music labels.
    <answer_labels>: the ground truth label for each query.
    For each query, compute the precision of the results of the K retrieved songs.
    Then average over queries.
    computed as: (number of match in top K) / K
    '''
    precisions= []
    for row,gt in zip(returned_labels, answer_labels):
        if not isinstance(gt, list):
            gt = [gt]
        row=row[:K]
        precisions.append(len(set(row) & set(gt))/K)
    return sum(precisions)/len(precisions)


def average_recall_at_K(returned_labels, answer_labels, K=K1):
    recalls = []
    for ranked, gt in zip(returned_labels, answer_labels):
        gt = gt if isinstance(gt, (list)) else [gt]
        retrieved_top_k = ranked[:K]
        hits = len(set(retrieved_top_k) & set(gt))
        recalls.append(hits / len(gt))
    return float(np.mean(recalls)) if recalls else 0.0


def average_mean_reciprocal_rank( returned_labels, answer_labels):
    '''
    <returned_labels>: number queries * K1. An np array of retrieved music labels.
    <answer_labels>: the ground truth label for each query.
    For each query, compute the rank of the ground truth song, if multiple ground
    truths exists, count the rank of the first ground truth song. Then take the
    reciprocal, then average over all queries.
    Compute as: sum(1/rank_query_i) / number_queries
    '''
    ranks= []
    for row,gt in zip(returned_labels, answer_labels):
        if not isinstance(gt, list):
            gt = [gt]

        # scan the row, the rank is the first  in row that is in the gt array
        found=False
        for i,retrieved in enumerate(row):
             if retrieved in gt:
                ranks.append(1/(i+1))
                found=True
                break
        if not found:
            ranks.append(1/N)
    return sum(ranks)/len(ranks)
```

## first stage search, second stage rerank, and evalution

```python
# ---------------------------------------- searching ----------------------------------------
tokenizer = AutoTokenizer.from_pretrained('intfloat/e5-large-v2')
model = AutoModel.from_pretrained('intfloat/e5-large-v2')
S, I = generate_score(search_index,tokenizer,model,len(queries),top_k=K1)

# ----------------------------------- reranking --------------------------------------
from sentence_transformers import CrossEncoder
import torch
DEVICE    = "cuda" if torch.cuda.is_available() else "cpu"
# Alibaba-NLP/gte-reranker-modernbert-base(decreases performance),
# jinaai/jina-reranker-v1-turbo-en (too bad), mixedbread-ai/mxbai-rerank-large-v1
# cross-encoder/ms-marco-MiniLM-L-6-v2
# BAAI/bge-reranker-base
reranker  = CrossEncoder("mixedbread-ai/mxbai-rerank-large-v1",device=DEVICE)
def rerank(
    queries,  # list [str]
    top_indices,    # I
    lyrics_db: list[str],
    top_k_final: int = 3
) -> list[list[tuple[int, float]]]:
    """
    <queries>: list of user queries.
    <top_indices>: The I being returned at the last stage, the indices for
    retrieved songs for each query.
    <lyrics_db>: the ith place is full lyric for song i.
    returns a list-of-lists which isfor each query, a list of tuples:
    (song_id, cross-enc score) sorted descendingly bt scores.
    """
    assert len(queries) == len(top_indices)

    flat_pairs, pair_owner = [], []
    for q_id, (q_text, cand_ids) in enumerate(zip(queries, top_indices)):
        for song_id in cand_ids:
            flat_pairs.append((q_text, lyrics_db[song_id]))
            pair_owner.append((q_id, song_id))

    # 1D array consisting of scores for each (query, candidate) pairs
    # in the order they're added
    flat_scores = reranker.predict(flat_pairs)

    # Re-assemble per query, each query is a spot in per_query, and is a list
    # that ranks by score: (song_id, score)
    per_query = [[] for _ in queries]
    for (q_id, song_id), score in zip(pair_owner, flat_scores):
        per_query[q_id].append((song_id, float(score)))
    for q_id in range(len(per_query)):
        per_query[q_id].sort(key=lambda x: x[1], reverse=True)
        per_query[q_id] = per_query[q_id][:top_k_final]
    return per_query

rerank_res = rerank(queries, I, search_index, 14)
new_I = []
for row in rerank_res:
  new_row =[]
  for el in row:
    new_row.append(int(el[0]))
  new_I.append(new_row)
I=new_I

# ---------------------------------------- evaluation ----------------------------------------
search_index_w_title_lower = list(map(str.lower, search_index_w_title))
answer_labels=convert_titles_to_labels(answers, search_index_w_title_lower)

precision_at_1=average_precision_at_K(I, answer_labels,K=1)
precision_at_3=average_precision_at_K(I, answer_labels,K=3)
precision_at_5=average_precision_at_K(I, answer_labels,K=5)
recall_at_1=average_recall_at_K(I, answer_labels,K=1)
recall_at_3=average_recall_at_K(I, answer_labels,K=3)
recall_at_5=average_recall_at_K(I, answer_labels,K=5)
# recall_at_14=average_recall_at_K(I, answer_labels,K=14)


MRR = average_mean_reciprocal_rank(I, answer_labels)
print("precision at 1", precision_at_1)
print("precision at 3", precision_at_3)
print("precision at 5", precision_at_5)

print("recall at 1", recall_at_1)
print("recall at 3", recall_at_3)
print("recall at 5", recall_at_5)
# print("recall at 14", recall_at_14)

print("mean reciprocal rank", MRR)
print("which translates to a mean rank of the first ground truth song:", 1/MRR)
```
